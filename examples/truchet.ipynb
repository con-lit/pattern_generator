```python
from truchet_tiling.truchet_pattern import generate
from truchet_tiling.commons.themes import BLUE_WAVES, RED_FLAMES, GREEN_FIELDS 

# helper to display bitmap images
def display_image(image_bytes):
  from IPython.display import display, Image as IPImage
  display(IPImage(data=image_bytes.getvalue()))

image = generate(
    width=700,
    height=500,
    colors = BLUE_WAVES,
    directions='vertical',
    arcs_probability=0.8,
    stroke_color=0xFFFFFF,
    stroke_width=14
)

display_image(image)
```
